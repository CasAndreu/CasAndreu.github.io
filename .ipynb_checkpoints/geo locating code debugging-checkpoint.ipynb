```r
# Function 2: "getMatches()" - extracts all reciprocal follower/following relationships for
# a given twitter user and saves IDs to a character vector
getMatches <- function(tweeter){
  # get the user's metadata
  try(user <- getUser(tweeter))
  if(!exists("user", mode = "S4")){
    stop(paste0(tweeter, " does not exist"))
  } else {
        # get friends and followers for user
    followers <- user$getFollowerIDs()
    friends <- user$getFriendIDs()
        # get friend-follower matches (reciprocal following)
    matches <- followers %in% friends
        # extract IDs for matches
    match_IDs <- followers[matches==TRUE]
    rm(user)
    return(match_IDs)
  }  
}
```

```r
# apiCheckMatches is a function which checks if the number of API calls have been depleted, and, if so, pauses
# execution until the API access limits have been reset. This newer version incorporates the time management function
# which was previously self-contained. 

apiCheckMatches <- function(){
    #get the current rate limits
  rate_limits.df <- getCurRateLimitInfo()
    #rate limits for the 2 key vars in question
  friends_api <- as.numeric(rate_limits.df$remaining[55])
  followers_api <- as.numeric(rate_limits.df$remaining[35])
    #other potential sources of API lockup
  if(length(warnings())>0){
    loop_warnings <- warnings()
  } else {
    loop_warnings <- "no warnings"
    names(loop_warnings) <- "no warnings"
  }
  is_done <- grepl("Rate limit encountered", names(loop_warnings))
    #conditions that have to be met for the process to sleep
  if(friends_api<1 | followers_api<1 | is_done==TRUE){
      #total time until API limits are replenished
    sleep_for <- as.numeric(difftime(rate_limits.df$reset[55], Sys.time(), tz = "UTC", units = "secs"))
    print(paste0("API limit reached. Wait ", ceiling(sleep_for), " seconds."))
    countDown(ceiling(sleep_for))
  }
}
```

```r
# Loop to test the new getMatches function - tweeters is only n=6, with 2 obviously non-existent twitter handles.

for(i in 1:length(tweeters)){
  start_time <- format(Sys.time(), "%H-%M-%S")
  try(matches <- getMatches(tweeters[i]))
  end_time <- format(Sys.time(), "%H-%M-%S")
  if(exists("matches")){
    match_freq <- length(matches)
    # add to list
    matches.ls <- c(matches.ls, list(matches))
    } else {
    match_freq <- NA
    matches.ls <- c(matches.ls, paste0(tweeters[i], "does not exist."))
  }
  # add to data frame
  newline <- data.frame(ID[i], tweeters[i], match_freq, start_time, end_time)
  names(newline) <- c("ID", "handle", "match_freq", "start_time", "end_time")
  match_frequencies.df <- rbind(match_frequencies.df, newline)
  rm(match_freq, newline, start_time, end_time)
  # api reset test
  apiCheckMatches()
  # writeout
  if(i %% 10 == 0){
    write.csv(match_frequencies.df, file = paste0("matches_out/iteration", "00", i, ".csv"))
    match_frequencies.df <- match_frequencies.df[-c(1:10),]
    save(matches.ls, file = paste0("matches_out/matches", "00", i, ".Rdata"))
    matches.ls <- matches.ls[-c(1:10)]
  }
  rm(matches)
}
```
